```python
# churn percentage
sns.countplot(x='Churn',data=df)
plt.title('Churn Percentage')
plt.savefig('Chrun Percentage.png')
plt.show()
```

```python
from google.colab import files
files.download("Chrun Percentage.png")
```

# Insights
---
* Most customers are staying with the company
* A smaller number of customers are leaving

```python
# Which type of customers leave more?
sns.countplot(x="Contract", hue='Churn',data=df)
plt.title('Contract Type')
plt.savefig('Contract Type.png')
plt.show()
```

```python
from google.colab import files
files.download("Contract Type.png")
```

 # Insights
---
* Customers with month-to-month contracts have higher churn
* Customers with long-term contracts(1 year, 2year) tend to stay more
* Longer commitment reduces customer churn


```python
# who leaves more
sns.histplot(data=df, x="MonthlyCharges",hue="Churn",bins=30)
plt.title('Monthly Charges')
plt.savefig('Monthly Charges.png')
plt.show()
```

```python
from google.colab import files
files.download("Monthly Charges.png")
```

# Insights
---
* Customers with higher monthly charges are more likely to churn
* Pricing may influencer customer decisions to leave

```python
# How long customers stay before leaving
sns.histplot(data=df,x='tenure',hue="Churn",bins=30)
plt.title('Tenure')
plt.savefig('Tenure.png')
plt.show()
```

```python
from google.colab import files
files.download("Tenure.png")
```

# Insights
---
* Customers with low tenure are more likely to churn
* Customers who stay longer tend to remain loyal
* Early Customers experience plays a crucial role in retention

# Recommendations
---
* Encourage customers to choose long-term contracts by offering discounts
* Provide better pricing plans or offers for high-paying customers
* Offer early support and engagement to reduce churn in initial months
